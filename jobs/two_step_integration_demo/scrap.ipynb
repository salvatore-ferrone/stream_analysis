```python
import gcs 
from gcs import path_handler as ph
import numpy as np 
import matplotlib.pyplot as plt
```

```python
GCname="Pal5"
MWpotential = "pouliasis2017pii"
```

```python
orbits = {}
for i in range(5):
    montecarlokey = "monte-carlo-{:03d}".format(i)
    t,v,y,z,vx,vy,vz = gcs.extractors.GCOrbits.extract_whole_orbit(ph.GC_orbits(MWpotential,GCname),montecarlokey)
    today_index = np.argmin(np.abs(t-0))
    orbits[montecarlokey] = np.array([t[0:today_index],v[0:today_index],y[0:today_index],z[0:today_index],vx[0:today_index],vy[0:today_index],vz[0:today_index]])
```

```python
fig,axis=plt.subplots(1,1,figsize=(6,6))
for key,orbit in orbits.items():
    axis.plot(orbit[1],orbit[2],label=key)
    axis.plot(orbit[1],orbit[2],label=key)
    axis.scatter(orbit[1][-1],orbit[2][-1],s=100,zorder=100,label=key,color="r")
axis.set_xlabel("x[kpc]")
axis.set_ylabel("y [kpc]")
```
